```python
from Crypto.Cipher import DES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes
import binascii

class DESExample:
    def __init__(self):
        self.raw_key = self.generate_symmetric_key()
        self.input_message = input("Enter message to encrypt: ")

        # Encrypt the input message
        encrypted_data = self.encrypt(self.raw_key, self.input_message.encode())
        print("Encrypted message (hex):", binascii.hexlify(encrypted_data).decode())

        # Decrypt the message
        decrypted_message = self.decrypt(self.raw_key, encrypted_data)
        print("Decrypted message:", decrypted_message.decode())


    def generate_symmetric_key(self):
        # Generate a random key for DES (must be 8 bytes)
        key = get_random_bytes(8)
        print("DES Symmetric key (hex):", binascii.hexlify(key).decode())
        return key

    def encrypt(self, key, clear):
        # Create a DES cipher object
        cipher = DES.new(key, DES.MODE_CBC)
        # Pad the plaintext
        padded_clear = pad(clear, DES.block_size)
        # Encrypt the padded plaintext
        iv = cipher.iv
        encrypted = iv + cipher.encrypt(padded_clear)  # Prepend IV to ciphertext
        return encrypted
    def decrypt(self, key, encrypted):
        # Extract the IV from the beginning of the encrypted data
        iv = encrypted[:DES.block_size]
        ciphertext = encrypted[DES.block_size:]
        # Create a DES cipher object with the same IV
        cipher = DES.new(key, DES.MODE_CBC, iv)
        # Decrypt the ciphertext
        decrypted_padded_clear = cipher.decrypt(ciphertext)
        # Unpad the decrypted plaintext
        return unpad(decrypted_padded_clear, DES.block_size)

if __name__ == "__main__":
    DESExample()
```

```python
import math
p, q, n, t = 0, 0, 0, 0
flag = 0
e, d, temp, m, en = [], [], [], [], []
msg = ""
def prime(pr):
    j = int(math.sqrt(pr))
    for i in range(2, j + 1):
        if pr % i == 0:
            return 0
    return 1
def ce():
    global e, d, t, p, q
    k = 0
    for i in range(2, t):
        if t % i == 0:
            continue
        flag = prime(i)
        if flag == 1 and i != p and i != q:
            e.append(i)
            flag = cd(e[k])
            if flag > 0:
                d.append(flag)
                k += 1
            if k == 99:
                break
def cd(x):
    k = 1
    while True:
        k = k + t
        if k % x == 0:
            return k // x
def encrypt():
    global msg, m, n, e, en, temp
    key = e[0]
    for i, char in enumerate(msg):
        pt = ord(char) - 96
        k = pow(pt, key, n)
        temp.append(k)
        ct = k + 96
        en.append(ct)
    en.append(-1)
    print("\nTHE ENCRYPTED MESSAGE IS")
    print(''.join(chr(c) for c in en[:-1]))
def decrypt():
    global temp, n, d, m
    key = d[0]
    i = 0
    while en[i] != -1:
        ct = temp[i]
        k = pow(ct, key, n)
        pt = k + 96
        m.append(pt)
        i += 1
    m.append(-1)
    print("\nTHE DECRYPTED MESSAGE IS")
    print(''.join(chr(c) for c in m[:-1]))

def main():
    global p, q, n, t, msg, m
    print("\nENTER FIRST PRIME NUMBER")
    p = int(input())
    flag = prime(p)
    if flag == 0:
        print("\nWRONG INPUT")
        return
    print("\nENTER ANOTHER PRIME NUMBER")
    q = int(input())
    flag = prime(q)
    if flag == 0 or p == q:
        print("\nWRONG INPUT")
        return
    print("\nENTER MESSAGE")
    msg = input()
    m = [ord(c) for c in msg]
    n = p * q
    t = (p - 1) * (q - 1)
    ce()
    print("\nPOSSIBLE VALUES OF e AND d ARE")
    for i in range(len(e)):
        print(f"\n{e[i]}\t{d[i]}")
    encrypt()
    decrypt()
if __name__ == "__main__":
    main()
```

```python
def power(a, b, mod):
    """Calculates (a^b) % mod using recursive exponentiation by squaring."""
    if b == 1:
        return a % mod
    t = power(a, b // 2, mod)
    if b % 2 == 0:
        return (t * t) % mod
    else:
        return ((t * t) % mod * a) % mod


def main():
    # Input values
    n = int(input("Enter the value of n (prime number): "))
    g = int(input("Enter the value of g (generator): "))
    x = int(input("Enter the value of x for the first person: "))
    a = power(g, x, n)
    y = int(input("Enter the value of y for the second person: "))
    b = power(g, y, n)

    # Calculate the shared keys
    key1 = power(b, x, n)  # Key for the first person
    key2 = power(a, y, n)  # Key for the second person
    print(f"Key for the first person is: {key1}")
    print(f"Key for the second person is: {key2}")

if __name__ == "__main__":
    main()
```

```python
import math
import struct
class MD5union:
    def __init__(self):
        self.w = 0
        self.b = bytearray(4)
    DigestArray = list[int]
    def func0(abcd: list[int]) -> int:
        return (abcd[1] & abcd[2]) | (~abcd[1] & abcd[3])
    def func1(abcd: list[int]) -> int:
        return (abcd[3] & abcd[1]) | (~abcd[3] & abcd[2])

    def func2(abcd: list[int]) -> int:
        return abcd[1] ^ abcd[2] ^ abcd[3]

    def func3(abcd: list[int]) -> int:
        return abcd[2] ^ (abcd[1] | ~abcd[3])

    @staticmethod
    def calctable(k: list[int]) -> list[int]:
        pwr = 2 ** 32
        for i in range(64):
            s = abs(math.sin(1 + i))
            k[i] = int(s * pwr)
        return k

    @staticmethod
    def rol(r: int, N: int) -> int:
        mask1 = (1 << N) - 1
        return ((r >> (32 - N)) & mask1) | ((r << N) & ~mask1)

def md5(msg: str, mlen: int) -> list[int]:
    h0 = [0x67452301, 0xEFCDAB89, 0x98BADCFE, 0x10325476]
    ff = [MD5union.func0, MD5union.func1, MD5union.func2, MD5union.func3] # Corrected function calls
    M = [1, 5, 3, 7]
    O = [0, 1, 5, 0]
    rot0 = [7, 12, 17, 22]
    rot1 = [5, 9, 14, 20]
    rot2 = [4, 11, 16, 23]
    rot3 = [6, 10, 15, 21]
    rots = [rot0, rot1, rot2, rot3]
    kspace = [0] * 64
    k = MD5union.calctable(kspace) # Corrected function call
    h = h0.copy()

    grps = 1 + (mlen + 8) // 64
    msg2 = bytearray(64 * grps)
    msg2[:mlen] = msg.encode()
    msg2[mlen] = 0x80

    u = struct.pack("<I", 8 * mlen)
    msg2[-8:-4] = u

    for grp in range(grps):
        mm = msg2[grp*64:(grp+1)*64]
        abcd = h.copy()
        for p in range(4):
            fctn = ff[p]
            rotn = rots[p]
            m, o = M[p], O[p]
            for q in range(16):
                g = (m * q + o) % 16
                f = abcd[1] + MD5union.rol(abcd[0] + fctn(abcd) + k[q + 16 * p] + # Corrected function calls
                    struct.unpack("<I", mm[4*g:4*g+4])[0], rotn[q % 4])
                abcd = [abcd[3], f, abcd[1], abcd[2]]
        h = [h[i] + abcd[i] for i in range(4)]

    return h

def main():
    msg = "The quick brown fox jumps over the lazy dog"
    d = md5(msg, len(msg))
    print("\t MD5 ENCRYPTION ALGORITHM IN PYTHON \n")
    print(f"Input String to be Encrypted using MD5 : \n\t{msg}")
    print("\nThe MD5 code for input string is: ")
    print("\t= 0x", end="")
    for val in d:
        print(f"{val:08x}", end="")
    print("\n")
    print("\t MD5 Encryption Successfully Completed!!!\n")
if __name__ == "__main__":
    main()
```

```python
import hashlib
def sha1():
    try:
        md = hashlib.sha1()
        print("Message digest object info: ")
        print(f" Algorithm = {md.name}")
        print(f" Provider = {hashlib.__name__}")
        print(f" ToString = {md}")

        input_str = ""
        md.update(input_str.encode())
        output = md.digest()
        print()
        print(f"SHA1(\"{input_str}\") = {bytes_to_hex(output)}")

        input_str = "abc"
        md = hashlib.sha1()
        md.update(input_str.encode())
        output = md.digest()
        print()
        print(f"SHA1(\"{input_str}\") = {bytes_to_hex(output)}")
        input_str = "abcdefghijklmnopqrstuvwxyz"
        md = hashlib.sha1()
        md.update(input_str.encode())
        output = md.digest()
        print()
        print(f"SHA1(\"{input_str}\") = {bytes_to_hex(output)}")
        print()
    except Exception as e:
        print(f"Exception: {e}")
def bytes_to_hex(b):
    return ''.join([f'{byte:02X}' for byte in b])
if __name__ == "__main__":
    sha1()
```
